```python
melbourne_airport['PressureDifferences'] = melbourne_airport['Pressure9am'] - melbourne_airport['Pressure3pm']
melbourne_airport['temp_differences'] = melbourne_airport['MaxTemp'] - melbourne_airport['MinTemp']
melbourne_airport
```

Heteroskedacity is the idea of inconsistency of variability of points on the scatterplot. For lower values of the predictor, there is great variability in the PressureDifferences response variable for lower values of the explanatory variable(the differences in temperature).


```python
sns.regplot('temp_differences', 'PressureDifferences', data = melbourne_airport)
```

```python
def standardize(column):
    return (column-np.mean(column))/np.std(column)
def correlation(column1, column2):
    return np.mean(standardize(melbourne_airport[column1])*standardize(melbourne_airport[column2]))
def slope(column1, column2):
    return correlation(column1, column2)*(np.std(melbourne_airport[column2])/np.std(melbourne_airport[column1]))
def y_int(column1, column2):
    return np.mean(melbourne_airport[column2]) - slope(column1, column2)*np.mean(melbourne_airport[column1])
```

```python
the_slope = slope('temp_differences', 'PressureDifferences')
the_yint = y_int('temp_differences', 'PressureDifferences')
the_correlation = correlation('temp_differences', 'PressureDifferences')
print(the_correlation)
#There is a weak positive association between temp_differences as the predictor and pressureDifferences as the response. 
the_slope, the_yint
```

```python
predictions_of_pressure = the_slope*melbourne_airport['temp_differences'] + the_yint
```

```python
mean_squared_error = np.mean((predictions_of_pressure - melbourne_airport['PressureDifferences'])**2)
mean_squared_error
```

```python
def normalize_temp_differences(row):
    return (row-np.mean(melbourne_airport['temp_differences']))/np.std(melbourne_airport['temp_differences'])
def normalize_pressure_differences(row):
    return (row-np.mean(melbourne_airport['PressureDifferences']))/np.std(melbourne_airport['PressureDifferences'])
melbourne_airport['temp_differences'] = melbourne_airport['temp_differences'].apply(normalize_temp_differences)
melbourne_airport['PressureDifferences'] = melbourne_airport['PressureDifferences'].apply(normalize_pressure_differences)
sns.scatterplot(x = melbourne_airport['temp_differences'], y = melbourne_airport['PressureDifferences'])
```

```python
#Model Development Phase. K nearest neighbors classifier
from sklearn.model_selection import train_test_split 
melbourne_airport = melbourne_airport.sample(len(melbourne_airport), replace = False)[['RainTomorrow','temp_differences', 'PressureDifferences']]
X = melbourne_airport.iloc[:, 1:]
y = melbourne_airport.iloc[:, 0]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=1) #3:2 train test split
```

**K Nearest Neighbors Classifier**

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(20)
neigh.fit(X_train, y_train)
print(neigh.predict(X_test))
```

```python
from sklearn.model_selection import cross_val_score
from sklearn import metrics
training_scores = cross_val_score(neigh, X_train, y_train, cv = 10)
testing_scores = cross_val_score(neigh, X_test, y_test, cv= 10)
print(training_scores)
print(testing_scores)
average_cv_training1 = np.mean(training_scores)
average_cv_testing1 = np.mean(testing_scores)
average_cv_training1, average_cv_testing1
```
